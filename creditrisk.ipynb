```python
#aggregate statistics DAYS_CREDIT column
def aggregate_days_credit(df, column_name):
    agg_df = df.groupby('SK_ID_CURR')[column_name].agg(['min', 'max', 'mean']).reset_index()
    agg_df.columns = ['SK_ID_CURR', f'{column_name}_MIN', f'{column_name}_MAX', f'{column_name}_AVG']
    return agg_df

days_credit_columns = ['DAYS_CREDIT', 'DAYS_CREDIT_ENDDATE', 'DAYS_CREDIT_UPDATE', 'DAYS_ENDDATE_FACT']

for col in days_credit_columns:
    agg_df = aggregate_days_credit(bureau, col)
    df = df.merge(agg_df, on='SK_ID_CURR', how='left')
```

```python
object_cols = previous_application.select_dtypes(include=['object']).columns

result_df = previous_application[['SK_ID_CURR']].drop_duplicates().set_index('SK_ID_CURR')

for col in object_cols:

    pivot_df = previous_application.pivot_table(index='SK_ID_CURR',
                                                columns=col,
                                                aggfunc='size',
                                                fill_value=0)

    pivot_df.columns = [f'{col}_{str(v)}_count' for v in pivot_df.columns]

    result_df = result_df.merge(pivot_df, left_index=True, right_index=True, how='left')


result_df.reset_index(inplace=True)

```

```python
# Identify amount-type columns (those starting with 'AMT_')
amount_cols = [col for col in previous_application.columns if col.startswith('AMT_')]

amt_sums = previous_application.groupby('SK_ID_CURR')[amount_cols].sum().reset_index()


amt_sums.columns = [f'{col}_prevSUM' if col != 'SK_ID_CURR' else col for col in amt_sums.columns]

df = df.merge(amt_sums, on='SK_ID_CURR', how='left')
```

```python
def get_part_of_day(hour):
    if 0 <= hour <= 5:
        return '00_05'
    elif 6 <= hour <= 11:
        return '06_11'
    elif 12 <= hour <= 17:
        return '12_17'
    elif 18 <= hour <= 23:
        return '18_23'

previous_application['PART_OF_DAY'] = previous_application['HOUR_APPR_PROCESS_START'].apply(get_part_of_day)


part_of_day_counts = previous_application.pivot_table(index='SK_ID_CURR',
                                                      columns='PART_OF_DAY',
                                                      aggfunc='size',
                                                      fill_value=0).reset_index()


part_of_day_counts.columns = ['SK_ID_CURR', 'HOUR_APPR_PROCESS_STARTT_00_05', 'HOUR_APPR_PROCESS_START_06_11', 'HOUR_APPR_PROCESS_START_12_17', 'HOUR_APPR_PROCESS_START_18_23']


df = df.merge(part_of_day_counts, on='SK_ID_CURR', how='left')

df.head()
```

```python
installments_payments['PAYMENT_INSTALMENT_RATIO'] = installments_payments['AMT_PAYMENT'] / installments_payments['AMT_INSTALMENT']

avg_payment_instalment_ratio = installments_payments.groupby('SK_ID_CURR')['PAYMENT_INSTALMENT_RATIO'].mean().reset_index()

avg_payment_instalment_ratio.rename(columns={'PAYMENT_INSTALMENT_RATIO': 'AVG_PAYMENT_INSTALMENT_RATIO'}, inplace=True)
avg_payment_instalment_ratio['AVG_PAYMENT_INSTALMENT_RATIO'].replace([np.inf, -np.inf], 0, inplace=True)


df = df.merge(avg_payment_instalment_ratio, on='SK_ID_CURR', how='left')

del avg_payment_instalment_ratio
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

non_numeric_columns = df.select_dtypes(include=['object']).columns

for col in non_numeric_columns:
    df[col] = label_encoder.fit_transform(df[col])
```

```python
from sklearn.model_selection import train_test_split
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, accuracy_score


y = df['TARGET']
X = df.drop(columns=['TARGET'])


neg, pos = y.value_counts()
scale = neg / pos

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

xgb_clf = XGBClassifier(
    objective='binary:logistic',
    random_state=42,
    n_estimators=100,
    learning_rate=0.1,
    max_depth=5,
    scale_pos_weight=10
)


xgb_clf.fit(
    X_train,
    y_train,
    eval_set=[(X_test, y_test)],
    verbose=True
)

y_pred = xgb_clf.predict(X_test)


print("Classification Report (Test):")
print(classification_report(y_test, y_pred))

print("Accuracy Score (Test):", accuracy_score(y_test, y_pred))


y_pred_proba = xgb_clf.predict_proba(X_test)[:, 1]

threshold = 0.45
y_pred_adjusted = (y_pred_proba >= threshold).astype(int)

print(f"Classification Report (Threshold = {threshold}):")
print(classification_report(y_test, y_pred_adjusted))

print(f"Accuracy Score (Threshold = {threshold}):", accuracy_score(y_test, y_pred_adjusted))
```

```python
results_df = pd.DataFrame({
    'true_labels': y_test,
    'predicted_proba': y_pred_proba
})

results_df.head()
```

```python

results_df['decile'] = pd.qcut(results_df['predicted_proba'], 10, labels=False) + 1


decile_analysis = results_df.groupby('decile').agg(
    total_loan_ids=('true_labels', 'count'),
    total_bad=('true_labels', 'sum')
).reset_index()

decile_analysis['bad_rate'] = decile_analysis['total_bad'] / decile_analysis['total_loan_ids']

# Sort by predicted probability to capture top 50 defaulters
results_df_sorted = results_df.sort_values(by='predicted_proba', ascending=False)
top_50_defaulters = results_df_sorted.head(50)

# Count how many defaulters are in the top 3 deciles
top_3_deciles_defaulters = top_50_defaulters[top_50_defaulters['decile'] >=7 ]['true_labels'].sum()


print(f"Number of defaulters in the top 3 deciles: {top_3_deciles_defaulters}")
print(decile_analysis)

```
